```python
import pandas as pd
import numpy as np
```

```python
path = '/kaggle/input/now-with-more-nutrients/cleaned_ingredients.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
df["skor_protein"] = df["Protein_g"].apply(lambda x: 0 if x < 10 else (1 if x <= 20 else 2))
df["skor_lemak"] = df["Fat_g"].apply(lambda x: 0 if x < 3 else (1 if x <= 10 else 2))
df['skor_carb'] = df["Carb_g"].apply(lambda x: 0 if x < 25 else (1 if x <= 50 else 2))
df["skor_gula"] = df["Sugar_g"].apply(lambda x: 0 if x < 1 else (1 if x <=5 else (2 if x <= 10 else 3)))
```

```python
# Jumlah total data
total_data = len(df)

# Jumlah setiap nilai 0, 1, 2, 3 yang harus sama
num_per_value = total_data // 4

# Jika total_data tidak dapat dibagi 4 secara merata, tambahkan sisanya
values = [0] * num_per_value + [1] * num_per_value + [2] * num_per_value + [3] * num_per_value

# Tambahkan nilai sisa secara merata
remainder = total_data % 4
for i in range(remainder):
    values.append(i)

# Acak urutan nilai dalam daftar
np.random.shuffle(values)

# Tambahkan nilai ke kolom 'user'
df['user'] = values
```

```python
df["total_skor"] = df["skor_protein"] + df["skor_lemak"] + df['skor_carb'] + df["skor_gula"] + df["user"]
```

```python
df["class"] = df["total_skor"].apply(lambda x: 0 if x <= 3 else(1 if x <= 6 else (2 if x <= 9 else 3)))
df
```

```python
# rename kolom
df.rename(columns={'Protein_g': 'protein', 'Fat_g': 'lemak', 'Carb_g': 'karbohidrat', 'Sugar_g': 'gula'}, inplace=True)
```
